#Read in CPLEX Solution files
And make data to style maps in the multi-objective visualization

```python
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
import os
```

```python
frontiers = ["NONE", "Ensemble_rcp85"]
```

```python
# get list of all variables using first solution
firstfile = "solutionFiles/NONE/NONE_0.sol"
variables = {"Frontier":[],"SolutionIndex":[]}
solTree = ET.parse(firstfile)
cpxSol = solTree.getroot()
# record all variables
for variable in cpxSol.iter("variable"):
    varname = variable.get("name")
    variables[varname] = []
# make dataframe of these values
dfsolutions = pd.DataFrame(variables)

# get list of all solution files
solFiles = []
solDir = "solutionFiles/NONE/"
for (dirpath, dirnames, filenames) in os.walk(solDir):
    solFiles.extend(filenames)
# extract all variable values from these solutions
for solfile in solFiles:
    curvars = variables
    solTree = ET.parse(firstfile)
    cpxSol = solTree.getroot()
    # record all variables
    for variable in cpxSol.iter("variable"):
        varname = variable.get("name")
        if varname in dfsolutions.columns:
            curvars[varname] = variable.get("value")
    dfsolutions = dfsolutions.append(curvars, ignore_index=True)

# write output to file
#dfsolutions.to_csv("solutions_allVarVals.csv", index=None)
```
